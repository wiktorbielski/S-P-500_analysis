```python
import pandas as pd
import matplotlib.pyplot as plt

top_10 = pd.read_csv('output/top_10_s&p500.csv', sep=";")

# Extract labels and values for the pie chart (use the numeric 'Weight' for plotting)
labels = top_10['Company']
sizes = top_10['Weight']  # Use the numeric 'Weight' values for the pie chart

# Plotting the pie chart with a smaller size and adjusted labels
plt.figure(figsize=(7,7))  # Smaller figure size
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors, labeldistance=1.1)
plt.title('Average Weight by Company', fontsize=14)

# Ensuring the title and labels fit within the plot area
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()  # Adjust layout to make room for labels and title

plt.show()
```
